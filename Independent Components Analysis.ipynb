```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import FastICA
# Generate synthetic data
np.random.seed(0)
n_samples = 1000
time = np.linspace(0, 8, n_samples)
s1 = np.sin(2 * time)  # Signal 1 : sinusoidal signal
s2 = np.sign(np.sin(3 * time))  # Signal 2 : square signal
s3 = np.random.randn(n_samples)  # Signal 3 : random noise
# Combine signals to create observed data
X = np.c_[s1, s2, s3]
# Perform ICA
ica = FastICA(n_components=3)
S_ = ica.fit_transform(X)  # Reconstruct independent components
# Plot the original signals and the independent components
plt.figure(figsize=(8, 6))
models = [s1, s2, s3, S_]
names = ['Signal 1 (Original)', 'Signal 2 (Original)', 'Signal 3 (Original)',
         'Independent Components (Recovered)']
colors = ['red', 'blue', 'green', 'orange']
for i, (model, name) in enumerate(zip(models, names), 1):
    plt.subplot(4, 1, i)
    plt.title(name)
    plt.plot(model, color=colors[i - 1])
plt.tight_layout()
plt.show()
```
